```python
import numpy as np
import matplotlib.pyplot as plt

# Define the IS_alpha function
def IS_alpha(y, alpha, l, u):
    sharpness = u - l
    overprediction = (2/alpha) * (l - y) * (y < l)
    underprediction = (2/alpha) * (y - u) * (y > u)
    return sharpness + overprediction + underprediction

# Parameters
alpha = 0.1
l = 25
u = 125

# Generate y values
y_values = np.linspace(0, 200, 400)
IS_values = IS_alpha(y_values, alpha, l, u)

# Plot the function
plt.figure(figsize=(12, 6))
plt.plot(y_values, IS_values, label=r'$\operatorname{IS}_{\alpha}(F,y)$', color='b')
plt.xlabel('y')
plt.ylabel(r'$\operatorname{IS}_{\alpha}(F,y)$')
plt.axvline(l, color='r', linestyle='--', label='lower bound l')
plt.axvline(u, color='g', linestyle='--', label='upper bound u')
plt.legend()
plt.grid(True)
plt.savefig("IS.pdf", bbox_inches='tight')
```
